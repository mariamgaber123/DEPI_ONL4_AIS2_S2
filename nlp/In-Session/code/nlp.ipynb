```python
import nltk
#This model is a pre-trained classifier that identifies named entities (such as people, organizations, locations, etc.) in text.
nltk.download('maxent_ne_chunker')
nltk.download('words')
from nltk import word_tokenize, pos_tag, ne_chunk
input_text = "Steve Jobs was the CEO of Apple Corp. in California."
tokens = word_tokenize(input_text)
pos_tags = pos_tag(tokens)
print(pos_tags)
ne_tree = ne_chunk(pos_tags)
print('ne_tree',ne_tree)
```

```python
named_entities =[]
for subtree in ne_tree.subtrees():
   if subtree.label() in ['PERSON', 'ORGANIZATION', 'LOCATION']:
      named_entity  = ' '.join(word for word ,tag in subtree.leaves())
      print(named_entity)
      named_entities.append((named_entity,subtree.label()))

print(named_entities)
```

```python
for word, tag in subtree.leaves():
    print(word)
```

```python
corpus = [
    "Thor ate pizza",
    "Loki is tall",
    "Loki is eating pizza"
]
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")

def preprocess(text):
   doc = nlp(text)
   filtered_tokens =[]
   for token in doc :
      if token.is_stop or token.is_punct:
         continue
      filtered_tokens.append(token.lemma_)
   return " ".join(filtered_tokens)
   
```

```python
preprocess("Thor ate pizza")
```

```python
corpus_processed = [
    preprocess(text) for text in corpus
]
corpus_processed
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sents=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
sents
```

```python
voc_size=10000
```

```python
onehot_repr =[one_hot(sent,voc_size) for sent in sents]
print(onehot_repr)
```

```python
import numpy as np
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
sent_len=8
embedded_docs =pad_sequences(onehot_repr,padding='pre',maxlen=sent_len)
print(embedded_docs)
```

```python
dim=10
model =Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_len))
model.build(input_shape=(32,sent_len))
model.compile('adam','mse')
model.summary()
```

```python
print(model.predict(embedded_docs))
```
