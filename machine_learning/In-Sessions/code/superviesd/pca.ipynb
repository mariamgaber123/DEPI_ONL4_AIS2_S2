```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.matrix([2.5,0.5,2.2, 1.9, 3.1, 2.3, 2, 1, 1.5, 1.1  ])
 
y = np.matrix([2.4,0.7,2.9,2.2,3.0,2.7,1.6,1.1,1.6,0.9])
original_data = np.r_[x,y]
original_data
```

```python
plt.scatter([x],[y])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
mean_x = np.mean(x)
mean_y = np.mean(y)
x = x-mean_x
y = y-mean_y
```

```python
plt.scatter([x],[y])
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
data = np.r_[x,y]
cov = np.matmul(data,data.T)/(10-1)
cov
```

```python
evals  ,evecs = np.linalg.eig(cov)
print(evals)
print(evecs)
```

```python
evecs = np.flip(evecs,axis=1)
evals = np.flip(evals)
print(evals)
print(evecs)
```

```python
print(np.dot(evecs[:,0].T,evecs[:,1]))
```

```python
plt.scatter([x],[y])
plt.plot([-evecs[0,1]*2,evecs[0,1]*2],[-evecs[1,1]*2,evecs[1,1]*2],'y')
plt.plot([-evecs[0,0]*2,evecs[0,0]*2],[-evecs[1,0]*2,evecs[1,0]*2],'r')
plt.axis('equal')
```
