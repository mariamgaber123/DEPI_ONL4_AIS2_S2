```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score,root_mean_squared_error
import warnings 
warnings.filterwarnings('ignore')
```

```python
data =pd.read_csv('D:/DEPI/Amit/Amit/machine_learning/In-Sessions/data_sets/Salary_Data.csv')
data.head()
```

```python
plt.figure(figsize=(3,3))
sns.pairplot(data,x_vars=["YearsExperience"],y_vars=["Salary"],size=7,kind="scatter")
plt.xlabel('years')
plt.ylabel('salary')
#plt.title('')
plt.show()
```

```python
x=data.iloc[:,:-1]
y=data.iloc[:,1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state= 10) 
```

```python
x_train
```

```python
linear_model=LinearRegression()
linear_model.fit(x_train,y_train)
```

```python
plt.scatter(x_train,y_train,color="red")
plt.plot(x_train,linear_model.predict(x_train),color="blue")
plt.title("Salary vs Experiance if years")
plt.xlabel("Years of Experiance")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(x_test,y_test,color="red")
plt.plot(x_test,linear_model.predict(x_test),color="blue")
plt.title("salary vs Experiance if years")
plt.xlabel("Years of Experiance")
plt.ylabel("Salary")
plt.show()
```

```python
y_pred=linear_model.predict(x_test)
y_pred
```

```python
y_test
```

```python
#calculate mean square error
mse = mean_squared_error(y_test,y_pred)
print(mse)
rms=root_mean_squared_error(y_test,y_pred)
print(rms)
r2score = r2_score(y_test,y_pred)
print(r2score)
```
