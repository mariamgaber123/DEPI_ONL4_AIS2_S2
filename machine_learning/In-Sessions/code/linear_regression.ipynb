```python
import numpy as np
import matplotlib.pyplot as plt
 
```

```python
x = np.array([1,2,3,4])
y = np.array([2,2.8,3.6,4.5])
```

```python
w = 0
b = 0
alpha =0.01
num_iteration = 20
```

```python
sse_value = []

for i in range(num_iteration):
   y_hat = w * x + b

   D_w = (2/4) * np.sum((y_hat-y)*x)
   D_b = (2/4) * np.sum(y_hat-y)

   w -=  alpha * D_w
   b -=  alpha * D_b
   
   sse = (np.sum((y_hat-y)**2))
   sse_value.append(sse)

   if (i+1)%20 == 0 :
      print(f"iteration {i+1}, sse{sse}")
      


```

```python
print(f"optemized param : slop = {w} , bais = {b}")
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(range(num_iteration) , sse_value , label='sse')
plt.xlabel("Iteration")
plt.ylabel("SSE")
plt.title("SSE over Iteration")
plt.legend()
plt.subplot(1,2,2)
plt.scatter(x,y,color="blue",label="Data point")
plt.plot(x,w*x+b,color="red" , label="Regression line")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Regression fit line")
plt.show()
```
