```python
from TaskLinearRegressionGD import LinearRegressionGD

import numpy as np
x = [50, 60, 70, 80, 90] 
y = [150, 180, 210, 240, 270] 

l=LinearRegressionGD(x,y)
#a very large learning rate
l.fit(0.001,100)
l.visualize()
print(l.predict(70))
print(l.predict([50,60]))
print(l.calc_MSE())
print(l.calc_RMSE())
print(l.calc_R2Score())
```

```python
l2=LinearRegressionGD(x,y)
#a very small learning rate 
l2.fit(0.000001,500)
l2.visualize()
print(l2.predict(70))
print(l2.predict([50,60]))
print(l2.calc_MSE())
print(l2.calc_RMSE())
print(l2.calc_R2Score())
```

```python
from TaskPolynomial import PolynomialRegression
```

```python
x = [1, 2, 3, 4, 5]
y = [3, 6, 11, 18, 27]  
```

```python
l5 = PolynomialRegression(x, y, 2)
l5.fit_transform(0.0001, 10000)

l5.visualize()

print("Prediction for 3:", l5.predict(3))
print("MSE:", l5.calc_MSE())
print("RMSE:", l5.calc_RMSE())
print("R2:", l5.calc_R2Score())
```

```python
import numpy as np
x = np.array(x)
x_range = np.linspace(x.min(), x.max(), 20).reshape(-1,1)
x_range_poly = l5.transform(x_range)
#print(x_range_poly)
y_range_pred = l5.predict(x_range_poly)
print(y_range_pred)
```
