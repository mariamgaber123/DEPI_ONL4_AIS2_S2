```python
#problem 3

class Solution(object):
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        result = []
        for i in range(len(nums)):
            for j in range(i+1,len(nums)):
               if nums[i] + nums[j] == target :
                    result.append(i)
                    result.append(j)  
                    break 
               
        return result

nums = [2,7,11,15]
target = 9
r = Solution()    
r.twoSum(nums,target)
```

```python
#problem 5
class Solution(object):
    def romanToInt(self,s):
        """
        :type s: str
        :rtype: int
        """
         # I             1
         # V             5
         # X             10
         # L             50
         # C             100
         # D             500
         # M             1000
        count = 0
        i = 0

        while i < len(s):
            if i + 1 < len(s) and s[i]=='I' and s[i+1]=='X' :
                       count+=9
                       i += 2 
            elif i + 1 < len(s) and s[i]=='I' and s[i+1]=='V' :
                       count+=4
                       i += 2
            
            elif i + 1 < len(s) and s[i]=='X' and s[i+1]=='L' :
                       count+=40
                       i += 2 
            elif i + 1 < len(s) and s[i]=='X' and s[i+1]=='C' :
                       count+=90
                       i += 2
            
            elif i + 1 < len(s) and s[i]=='C' and s[i+1]=='D' :
                       count+=400 
                       i += 2
            elif i + 1 < len(s) and s[i]=='C' and s[i+1]=='M' :
                       count+=900
                       i += 2

            elif   s[i]=='X' :
                       count+=10
                       i += 1
            elif   s[i]=='I' :
                       count+=1
                       i += 1 
            elif   s[i]=='C' :
                       count+=100
                       i += 1             
            elif   s[i]=='V' :
                       count+=5
                       i += 1    
            elif   s[i]=='L' :
                       count+=50
                       i += 1                      
            elif   s[i]=='D' :
                       count+=500
                       i += 1    
            elif   s[i]=='M' :
                       count+=1000
                       i += 1      

        return   count                        
           

r = Solution()    
r.romanToInt("IIIV")   
```

```python
#problem 6
class Solution(object):
    def searchInsert(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        for i in range(len(nums)):  
                if  target  == nums[i]:
                    return i  

                elif  target  < nums[i]:
                    return i  
        return len(nums)

r = Solution()    
r.searchInsert([1,3,5,6],5)          
```
