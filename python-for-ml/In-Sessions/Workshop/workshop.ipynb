```python
#factorial
def factorial_1(num):
   if num==0:
      return 1
   if num==1:
      return 1
   return num*factorial_1(num-1)

factorial_1(5)
```

```python
#prime number
def is_prime(num):
   for i in range(2,num):
      if num % i==0:
         return False
   return True

is_prime(3)   
```

```python
# function Reverse
def reverse_txt(t):
   """
   Docstring for reverse_txt
   :param t: take string 
   :return :reversed of t
   """
   split_txt=t.split(" ")
   new_list=split_txt[::-1]
   result=" ".join(new_list)
   return result
 
reverse_txt("Depi Welcome")
```

```python
#Random password [a-z,!@#$%&^*.!]
import random
def generate_rand_code():
   """ 
     this function generate randome password contain 8 characters
     return: generated  password
     rtyp:str

   """
   characters = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ1234567890!@#$%^&*"
   passowrd = ""
   lenght=8
   for _ in range(lenght):
      passowrd+=random.choice(characters)

   return passowrd

print(f"Generate_Password: {generate_rand_code()}")
```
