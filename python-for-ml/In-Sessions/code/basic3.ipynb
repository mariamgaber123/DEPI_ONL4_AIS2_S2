# Functions

```python
# function and documentation
def user_info(name :str, age:int):
   """
   this function take name and age from user and return again 
   :param_1 : take the name of user from input user interface
   type param_1 : str
   :param_2 : take the age of user from input user interface
   type param_2 : int

   return : this function will print the name and the age 
   rtyp: str , int 
   """
   return f"my name is {name},my age is {age}"

user_info("mariam",20)
```

```python
#positional argument function
print(f"positional{user_info(20,"mariam")}")

#keyword argument function
print(f"keyword{user_info(age=20,name="mariam")}")

#default argument function
def user_info(name ="test", age=20):
       return f"my name is {name},my age is {age}"

#varadic argument function
def user_info(*name):
      """
      """
      print(name)
      print(type(name))


user_info("depi","python",[1,2,3])
#kwarge argument function
def user_info(**name):
      """
      """
      print(name)
      print(type(name))
user_info(key1=20,key2=3)
```
