```python
#Abstraction
from abc import ABC,abstractmethod

########
class Animal(ABC):

   @abstractmethod
   def move(self):
      pass

   @abstractmethod
   def sound(self):
      pass

#######
class Bird(Animal):
    
    def move(self):
      print("move from bird")


#######
class Cat(Animal):
   #  pass
    def move(self):
       print("move from cat")

    def sound(self):
       print("cat sound meeaww")   
```

```python
#Abstraction
a = Cat()
a.move()
```
