```python
#Task-1: 
email="mariam@gmail.com"
if(email.count('@')==1 and email.count('.')>=1):
   indexof_at= email.index("@")
   indexof_dot= email.index(".")
   Username=email[:indexof_at]
   Domain=email[indexof_at +1:indexof_dot]
   print("Username is ",Username)
   print("Domain is ",Domain)

   if(email.endswith(".com")):
      print( "Commercial Domain")
   elif(email.endswith(".edu")):
      print( "Educational Domain")
   else:
      print( "Othor Domain")
else:
   print("Invalid email")   
```

```python
#Task-2: 
Encode_message="###!!@mocleW EPGTQ!!!6789"
word=Encode_message.removeprefix("###!!@").removesuffix("!!!6789")
word_1=word.split()[0]
word_2=word.split()[1]
l=list(word_1)
l.reverse()
word_1=''.join(l)
result=word_1+word_2
result=result.lower()
word_2=result.removeprefix("welcome").upper()
word_1=result.removesuffix("pgtq").capitalize()
result=word_1+" "+word_2
print(result)
```
