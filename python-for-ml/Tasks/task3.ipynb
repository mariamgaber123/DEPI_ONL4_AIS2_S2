```python
class Fibonacci():
    """
    Fibonacci class generates a Fibonacci sequence.

    Attributes:
    n (int): Number of Fibonacci terms to generate.
    """

    def __init__(self, n: int):
        """
        Constructor to initialize the number of terms.

        Parameters:
        n (int): Number of Fibonacci elements.
        """
        self.n = n

    def generate_sequence(self):
        """
        Generates Fibonacci sequence from 0 up to n terms.

        Returns:
        list: Fibonacci sequence as a list.
        """
        seq = []
        num1, num2 = 0, 1

        for _ in range(self.n):
            seq.append(num1)
            num1, num2 = num2, num1 + num2
        return seq
    
#object      
fibo_num=Fibonacci(10)    
print(fibo_num.generate_sequence())
```

```python
class RemoveInSet():
    """
    RemoveInSet class remove element from set.

    Attributes:
    n (int): The Number want to remove from set .
    s (set):The set .
    """

    def __init__(self, s:set,n:int):
        """
        Constructor to initialize the number want to remove  .

        Parameters:
        n (int):The Number want to remove.
        """
        self.n = n
        self.s = s

    def remove_num(self):
        """
        Remove number from set .

        Returns:
        set: set after remove.
        """
        self.s.remove(self.n)    
        return self.s
    

#object    
my_set={1,2,3,4,5,6}
set_to_remove=RemoveInSet(my_set,4)    
print(set_to_remove.remove_num())
```

```python
class RemoveMultipleFromSet():
    """
    RemoveMultipleFromSet class removes multiple numbers from a set.

    Attributes:
    s (set): The set.
    numbers (list): List of numbers to remove from the set.
    """

    def __init__(self, s: set, numbers: list):
        """
        Constructor to initialize the set and numbers.

        Parameters:
        s (set): The set.
        numbers (list): Numbers to remove.
        """
        self.s = s
        self.numbers = numbers

    def remove_numbers(self):
        """
        Remove multiple numbers from the set.

        Returns:
        set: Set after removing numbers.
        """
        for i in self.numbers:
            self.s.remove(i)
        return self.s


# object
my_set = {1, 2, 3, 4, 5}
numbers = input("Enter numbers you want to remove: ")
list_of_num = numbers.split()

# convert to int
new_list = []
for i in list_of_num:
    i = int(i)
    new_list.append(i)

remove_obj = RemoveMultipleFromSet(my_set, new_list)
print(remove_obj.remove_numbers())
```

```python
class FindIndexInTuple():
    """
    FindIndexInTuple class finds the index of an element in a tuple.

    Attributes:
    t (tuple): The tuple.
    value (str): The value to find its index.
    """

    def __init__(self, t: tuple, value: str):
        """
        Constructor to initialize the tuple and the value.

        Parameters:
        t (tuple): The tuple.
        value (str): The value to find.
        """
        self.t = t
        self.value = value

    def find_index(self):
        """
        Find index of the value in the tuple.

        Returns:
        int: Index of the value.
        """
        return self.t.index(self.value)


# object
my_tuple = ('apple', 'banana', 'orange', 'grape', 'banana')
index_obj = FindIndexInTuple(my_tuple, 'orange')
print(index_obj.find_index())
```

```python
class SplitSentence():
    """
    SplitSentence class splits a sentence into words.

    Attributes:
    sentence (str): The sentence to split.
    """

    def __init__(self, sentence: str):
        """
        Constructor to initialize the sentence.

        Parameters:
        sentence (str): The sentence to split.
        """
        self.sentence = sentence

    def split_words(self):
        """
        Split the sentence into words.

        Returns:
        list: List of words in the sentence.
        """
        return self.sentence.split()


# object
sentence = input("Enter sentence: ")
split_obj = SplitSentence(sentence)
print(split_obj.split_words())
```

```python
class CheckPalindrome():
    """
    CheckPalindrome class checks whether a word is a palindrome or not.

    Attributes:
    word (str): The word to check.
    """

    def __init__(self, word: str):
        """
        Constructor to initialize the word.

        Parameters:
        word (str): The word to check.
        """
        self.word = word

    def is_palindrome(self):
        """
        Check if the word is a palindrome.

        Returns:
        str: 'Palindrome' if the word is a palindrome, else 'Not a Palindrome'.
        """
        rword = reversed(self.word)
        plind = True
        for i, j in zip(self.word, rword):
            if i != j:
                plind = False
                break
        if plind:
            return "Palindrome"
        else:
            return "Not a Palindrome"


# object
word = input("Enter word: ")
palindrome_obj = CheckPalindrome(word)
print(palindrome_obj.is_palindrome())
 
```

```python
class Factorial():
    """
    Factorial class calculates the factorial of a number.

    Attributes:
    n (int): The number to calculate factorial for.
    """

    def __init__(self, n: int):
        """
        Constructor to initialize the number.

        Parameters:
        n (int): The number to calculate factorial for.
        """
        self.n = n

    def calculate(self):
        """
        Calculate factorial of the number using recursion.

        Returns:
        int: Factorial of the number.
        """
        if self.n == 0 or self.n == 1:
            return 1
        else:
            return self.n * Factorial(self.n - 1).calculate()


# object
num = int(input("Enter number: "))
fact_obj = Factorial(num)
print(fact_obj.calculate())
```

```python
class Squares():
    """
    Squares class generates a list of squares of numbers in a given range.

    Attributes:
    start (int): Starting number of the range.
    end (int): Ending number of the range (inclusive).
    """

    def __init__(self, start: int, end: int):
        """
        Constructor to initialize the range.

        Parameters:
        start (int): Starting number.
        end (int): Ending number (inclusive).
        """
        self.start = start
        self.end = end

    def generate_squares(self):
        """
        Generate list of squares from start to end.

        Returns:
        list: List of squares of numbers.
        """
        squares = []
        for i in range(self.start, self.end + 1):
            squares.append(i * i)
        return squares


# object
squares_obj = Squares(1, 10)
print(squares_obj.generate_squares())
```

```python
class SumEvenNumbers():
    """
    SumEvenNumbers class calculates the sum of even numbers in a list.

    Attributes:
    numbers (list): List of numbers to calculate sum of even numbers.
    """

    def __init__(self, numbers: list):
        """
        Constructor to initialize the list of numbers.

        Parameters:
        numbers (list): List of numbers.
        """
        self.numbers = numbers

    def sum_even(self):
        """
        Calculate sum of even numbers in the list.

        Returns:
        int: Sum of even numbers.
        """
        total = 0
        for i in self.numbers:
            if i % 2 == 0:
                total += i
        return total


# object
nums = [1, 2, 3, 4, 5]
sum_obj = SumEvenNumbers(nums)
print(sum_obj.sum_even())
```

```python
class ReverseList():
    """
    ReverseList class reverses a list of numbers.

    Attributes:
    lst (list): List of numbers to reverse.
    """

    def __init__(self, lst: list):
        """
        Constructor to initialize the list.

        Parameters:
        lst (list): List of numbers.
        """
        self.lst = lst

    def reverse_list(self):
        """
        Reverse the list of numbers.

        Returns:
        list: Reversed list.
        """
        self.lst.reverse()
        return self.lst


# object
my_list = [1, 2, 3]
reverse_obj = ReverseList(my_list)
print(reverse_obj.reverse_list())
```
