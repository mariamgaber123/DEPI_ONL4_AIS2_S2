```python
import os 
import random

def makefiles(n:int):
   """ 
   this function open directory and create number of files in this directory
   
   :param n: take number of files you want to  create 
   :type n: int

   """
   dir="D:/DEPI/Amit/Amit/python-for-ml/My-Assignment/dir_of_assignment"
   os.makedirs(dir ,exist_ok=True)
   for i in range(n):
      innerpath= os.path.join(dir,f"file_{i+1}.txt")
      with open(innerpath, "w") as f:
            f.write(f"This is file {i+1}\n")

   files_b=os.listdir(dir)  
   print(f"lenght of files before delete is {len(files_b)}")    


   half=len(files_b)//2
   deleted=random.sample(files_b,half)
   for i in deleted:
       os.remove(os.path.join(dir,i))

   files_a=os.listdir(dir)  
   print(f"lenght of files after delete is {len(files_a)}")    

##############################################


makefiles(10)
```
