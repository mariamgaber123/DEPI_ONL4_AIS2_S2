```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer

```

```python
data = load_breast_cancer()
```

```python
df = pd.DataFrame(data=data['data'],columns=data['feature_names'])
df['target'] =data['target']
df.head()
```

```python
df['target'].value_counts()
```

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
def sigmoid(x):
   return 1/(1+np.exp(-x))
```

```python
def update_weights(weights,x,z,y,lr):
   m = x.shape[0]
   dw = (1/m) *np.dot(x.T,(z-y))
   weights = weights-lr*dw
   return weights
```

```python
def update_bais(bais,x,z,y,lr):
    m = x.shape[0]
    db = 1/m *np.sum(z-y)
    bais = bais-lr*db
    return bais
```

```python
def perceptron(x,weights,bais):
   z = np.dot(x,weights)+bais
   y_pred = sigmoid(z)
   y_p =(y_pred >= 0.5).astype(int)
   return y_p
```

```python
def accuracy(y_pred,y_test):
   accuracy = np.sum(y_pred==y_test)/len(y_test)
   return accuracy
```

```python
lr= 0.001
weights=np.zeros(x_train.shape[1])
bais = 0
for i in range(100):
   pred = perceptron(x_train,weights,bais)
   weights=update_weights(weights,x_train,pred,y_train,lr)
   bais=update_bais(bais,x_train,pred,y_train,lr)
   
```

```python
y_pred = perceptron(x_test,weights,bais)
print(accuracy(y_pred,y_test))
```
