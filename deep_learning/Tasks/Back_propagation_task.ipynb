```python
import numpy as np
import pandas as pd
```

```python
def sigmoid(x):
   return 1/(1+np.exp(-x))
```

```python
def sigmoid_derivative(x):
   return x *(1-x)
```

```python
# inputs and targets and initial weigths
x1 = 0.05
x2 = 0.10

target_o1 = 0.01
target_o2 = 0.99

w1, w2, w3, w4 = 0.15, 0.20, 0.25, 0.30
w5, w6, w7, w8 = 0.40, 0.45, 0.50, 0.55

b_h = 0.35
b_o = 0.60

lr = 0.5
```

```python
for epoch in range(10000):
   h1_net = x1*w1 + x2*w2 +b_h
   h2_net = x1*w3 + x2*w4 +b_h

   h1 = sigmoid(h1_net)
   h2 = sigmoid(h2_net)
   
   o1_net = h1*w5 + h2*w6 +b_o
   o2_net = h1*w7 + h2*w8 +b_o

   o1 = sigmoid(o1_net)
   o2 = sigmoid(o2_net)

   ##calculate Error
   E_o1 = 0.5 * (target_o1 -o1)**2
   E_o2 = 0.5 * (target_o2 -o2)**2

   E_t = E_o1 + E_o2 #total_error 

   d_o1 =(o1 - target_o1) * sigmoid_derivative(o1)
   d_o2 =(o2 - target_o2) * sigmoid_derivative(o2)

   dw5 = d_o1 * h1
   dw6 = d_o1 * h1
   dw7 = d_o2 * h2
   dw8 = d_o2 * h2

   d_h1 =(d_o1*w5 + d_o2*w7) * sigmoid_derivative(h1)
   d_h2 =(d_o1*w6 + d_o2*w8) * sigmoid_derivative(h2)

   dw1 = d_h1 * x1
   dw2 = d_h1 * x1
   dw3 = d_h2 * x2
   dw4 = d_h2 * x2
   
   #update weights

   w1 -= lr * dw1
   w2 -= lr * dw2
   w3 -= lr * dw3
   w4 -= lr * dw4
   w5 -= lr * dw5
   w6 -= lr * dw6
   w7 -= lr * dw7
   w8 -= lr * dw8
   
   if epoch % 10000 == 0:
        print(f"Epoch {epoch}, Error: {E_t}")

        
print("\nFinal output after training:")

h1 = sigmoid(x1*w1 + x2*w2 + b_h)
h2 = sigmoid(x1*w3 + x2*w4 + b_h)

o1 = sigmoid(h1*w5 + h2*w6 + b_o)
o2 = sigmoid(h1*w7 + h2*w8 + b_o)

print("o1 =", o1)
print("o2 =", o2)
print("NEW Weights")
print(w1, w2, w3, w4,w5, w6, w7, w8)
```
